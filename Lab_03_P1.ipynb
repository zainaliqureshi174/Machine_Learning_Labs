```python
# Importing necessary Libraries
import pandas as pd
import numpy as np
```

```python
# Using IRIS Dataset
filename = "Iris.csv"
Data = pd.read_csv(filename)
Data
```

## Sepal_Length

```python
Sepal_Length = Data.get('SepalLengthCm').to_numpy()
Sepal_Length
```

```python
Sorted = np.sort(Sepal_Length)
print(Sorted)
Mean = round(np.mean(Sepal_Length), 1)
Min = round(np.min(Sepal_Length), 1)
Max = round(np.max(Sepal_Length), 1)
print()
print(f"Minimum value: {Min}")
print(f"Maximum value: {Max}")
print(f"Mean of Data: {Mean}\n")

Range = (Max-Min)/3

SL_Small = []
S_Range = Min+Range
SL_Medium = []
M_Range = S_Range+Range
SL_Large = []
L_Range = M_Range+Range

print(f"Small Ranges: {Min}-{S_Range}.")
for i in range(150):
    if Sorted[i] < S_Range:
        SL_Small.append(Sorted[i])
print(SL_Small)
print(f"\nMedium Ranges: {S_Range}-{M_Range}.")
for i in range(150):
    if Sorted[i] < M_Range and Sorted[i] >= S_Range:
        SL_Medium.append(Sorted[i])
print(SL_Medium)
print(f"\nLarge Ranges: {L_Range}-{Max}.")
for i in range(150):
    if Sorted[i] <= Max and Sorted[i] >= M_Range:
        SL_Large.append(Sorted[i])
print(SL_Large)
SL = []

for i in range(150):
        if Sepal_Length[i] in SL_Small:
            SL.append("S")
        elif Sepal_Length[i] in SL_Medium:
            SL.append("M")
        elif Sepal_Length[i] in SL_Large:
            SL.append("L")
        
print(SL)

Data.insert(2, "SL", SL, True)
Data
```

## Calculating Probability

```python
category = input("Enter Feature Category: ")
value = float(input(f"Enter {category}: "))
flower = input("Enter the Flower: ")

if category == 'SL':
    if value in SL_Small:
        value = 'S'
    elif value in SL_Medium:
        value = 'M'
    elif value in SL_Large:
        value = 'L'
elif category == 'SW':
    if value in SW_Small:
        value = 'S'
    elif value in SW_Medium:
        value = 'M'
    elif value in SW_Large:
        value = 'L'
elif category == 'PL':
    if value in PL_Small:
        value = 'S'
    elif value in PL_Medium:
        value = 'M'
    elif value in PL_Large:
        value = 'L'
elif category == 'PW':
    if value in PW_Small:
        value = 'S'
    elif value in PW_Medium:
        value = 'M'
    elif value in PW_Large:
        value = 'L'

count = Data[(Data[category] == value) & (Data['Species'] == flower)].shape[0]
prob = count/150

print(f"Probability of {flower} with {category} {value} is: {round(prob, 2)}.")
```
